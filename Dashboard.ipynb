```python
%sql
SELECT
    fs.IdJugador,
    dj.Apellido,
    dj.Nombre,
    COUNT(CASE WHEN TiempoJuego > 0 THEN 1 END) AS partidos,
    ROUND(SUM(TiempoJuego), 2) AS minutos_totales,
    ROUND(SUM(TiempoJuego) / NULLIF(COUNT(CASE WHEN TiempoJuego > 0 THEN 1 END), 0), 2) AS minutos_x_partido,
    SUM(Puntos) AS puntos_totales,
    ROUND(SUM(Puntos) / NULLIF(COUNT(CASE WHEN TiempoJuego > 0 THEN 1 END), 0), 2) AS puntos_x_partido,
    SUM(asist) AS asistencias_totales,
    SUM(rebs_totales) AS rebotes_totales,
    SUM(recups) AS recuperaciones_totales,
    SUM(perdidas) AS perdidas_totales,
    SUM(tapas_realizadas) AS tapas_totales,
    SUM(tiros_2_totales) AS tiros_2_totales,
    SUM(tiros_2_aciertos) AS tiros_2_aciertos,
    ROUND(SUM(tiros_2_aciertos) / NULLIF(SUM(tiros_2_totales), 0) * 100, 2) AS pctg_2,
    SUM(tiros_3_totales) AS tiros_3_totales,
    SUM(tiros_3_aciertos) AS tiros_3_aciertos,
    ROUND(SUM(tiros_3_aciertos) / NULLIF(SUM(tiros_3_totales), 0) * 100, 2) AS pctg_3,
    SUM(tiros_libres_totales) AS tiros_libres_totales,
    SUM(tiros_libres_aciertos) AS tiros_libres_aciertos,
    ROUND(SUM(tiros_libres_aciertos) / NULLIF(SUM(tiros_libres_totales), 0) * 100, 2) AS pctg_libres,
    ROUND(SUM(Puntos) / NULLIF(2 * (SUM(tiros_2_totales) + SUM(tiros_3_totales) + 0.44 * SUM(tiros_libres_totales)), 0), 2) AS ts,
    ROUND((SUM(tiros_2_aciertos) + 1.5 * SUM(tiros_3_aciertos)) / NULLIF(SUM(tiros_2_totales) + SUM(tiros_3_totales), 0), 2) AS efg,
    ROUND(AVG(uso_pctg), 2) AS uso_pctg
FROM laliga.gold.fact_stats fs
JOIN laliga.gold.dim_jugadores dj 
ON fs.IdJugador = dj.IdJugador
WHERE fs.IdEquipo = '89192'
GROUP BY fs.IdJugador, dj.Apellido, dj.Nombre
ORDER BY puntos_totales DESC
```
