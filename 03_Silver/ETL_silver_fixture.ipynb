```python
%sql

MERGE INTO laliga.silver.fixture AS target
USING(
  SELECT
    id_partido,
    url, 
    SPLIT(SPLIT(url, '/')[size(SPLIT(url, '/')) - 1], '-vs-')[0] AS local_team,
    SPLIT(SPLIT(url, '/')[size(SPLIT(url, '/')) - 1], '-vs-')[1] AS visit_team,
    date AS match_date,
    'laliga.bronze.fixture' AS _source_table,
    CURRENT_TIMESTAMP() AS _processing_timestamp
  FROM laliga.bronze.fixture
  ) AS source
ON target.id_partido = source.id_partido
WHEN NOT MATCHED THEN INSERT *
```
