```python
%sql
MERGE INTO laliga.silver.players_stats AS target
USING(
  SELECT 
    id_partido,
    CAST(CAST(IdJugador AS BIGINT) AS STRING) AS IdJugador, -- transformando el valor a string
    CAST(IdClub AS STRING) AS IdClub,
    CAST(IdEquipo AS STRING) AS IdEquipo,
    Dorsal,
    SPLIT(NombreCompleto, ',')[0] AS Apellido, -- split de valor para tomar el de la columna
    SPLIT(NombreCompleto, ',')[1] AS Nombre,
    CASE WHEN CincoInicial = True THEN 1 ELSE 0 END AS CincoInicial, -- Pasamos el valor boolean a int y de true/false a 1/0
    ROUND(CAST(SPLIT(TiempoJuego, ':')[0] AS INT) + CAST(SPLIT(TiempoJuego, ':')[1] AS INT) / 60, 2) AS TiempoJuego,
    Puntos,
    TirosDos.Aciertos AS tiros_2p_aciertos,
    TirosDos.Fallados AS tiros_2p_fallados,
    TirosDos.Totales AS tiros_2p_totales,
    TirosDos.Porcentaje AS tiros_2p_pctg,
    TirosTres.Aciertos AS tiros_3p_aciertos,
    TirosTres.Fallados AS tiros_3p_fallados,
    TirosTres.Totales AS tiros_3p_totales,
    TirosTres.Porcentaje AS tiros_3p_pctg,
    TirosLibres.Aciertos AS libres_aciertos,
    TirosLibres.Fallados AS libres_fallados,
    TirosLibres.Totales AS libres_totales,
    TirosLibres.Porcentaje AS libres_pctg,
    ReboteDefensivo,
    ReboteOfensivo,
    RebotesTotales,
    Asistencias,
    Recuperaciones,
    Perdidas,
    TaponCometido,
    TaponRecibido,
    FaltaCometida,
    FaltaRecibida,
    'laliga.bronze.partidos_stats' AS _source_table,
    CURRENT_TIMESTAMP() AS _processing_timestamp
  FROM laliga.bronze.partidos_stats
  WHERE IdClub != 0
  AND IdJugador IS NOT NULL
) AS source
ON target.id_partido = source.id_partido
AND target.IdJugador = source.IdJugador
WHEN NOT MATCHED THEN INSERT *
```
