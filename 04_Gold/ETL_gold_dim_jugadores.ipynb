```python
%sql
MERGE INTO laliga.gold.dim_jugadores AS target
USING(
  SELECT
     DISTINCT (IdJugador),
    IdClub,
    IdEquipo,
    Dorsal,
    Apellido,
    Nombre,
    'laliga.silver.players_stats' AS _source_table,
    CURRENT_TIMESTAMP() AS _processing_timestamp
  FROM laliga.silver.players_stats
  ) AS source
ON target.IdJugador = source.IdJugador
WHEN NOT MATCHED THEN INSERT *
```
