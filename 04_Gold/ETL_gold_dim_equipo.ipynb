```python
%sql

MERGE INTO laliga.gold.dim_equipo AS target
USING(
  SELECT
    DISTINCT (IdEquipo),
    IdClub,
    CASE WHEN IdEquipo = '89192' THEN 'san-lorenzo' ELSE 'unknown' END AS NombreEquipo,
    'laliga.silver.players_stats' AS _source_table,
    CURRENT_TIMESTAMP() AS _processing_timestamp
  FROM laliga.silver.players_stats 
) AS source
ON target.IdEquipo = source.IdEquipo
WHEN NOT MATCHED THEN INSERT *
```
