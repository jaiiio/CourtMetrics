```python
%sql
MERGE INTO laliga.gold.dim_partidos AS target
USING(
  SELECT
    id_partido,
    url,
    local_team,
    visit_team,
    match_date,
    'laliga.silver.fixture' AS _source_table,
    CURRENT_TIMESTAMP() AS _processing_timestamp
  FROM laliga.silver.fixture
) AS source
ON target.id_partido = source.id_partido
WHEN NOT MATCHED THEN INSERT *
```
