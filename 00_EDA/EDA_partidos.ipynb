### Partidos como parte de la plantilla

```python
%sql

SELECT 
  IdJugador,
  Nombre,
  COUNT(*) AS partidos_citado
FROM laliga.bronze.partidos_stats
WHERE IdEquipo = 89192
GROUP BY IdJugador, Nombre 
ORDER BY partidos_citado DESC
```
